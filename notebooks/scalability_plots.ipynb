```python
def extract_mfos_run(grid_size, seed):
    mfos_run_log_dir = f'/Users/miladaghajohari/PycharmProjects/loqa/league/checkpoints/mfos_seeds/camera_ready_self_{grid_size}x{grid_size}_{seed}'
    max_i = 10000
    MFOS_NUM_STEPS = 256 # WARNING: this is hardcoded
    metrics = []
    # instead of playing smart moves, I just iterate over all possible files
    # I didn't want to write complex code to extract file indices, sorting them, etc
    for i in range(max_i):
        # if exists, load the file, use the path module
        file = f'{mfos_run_log_dir}/out_{i}.json'
        if not os.path.exists(file):
            continue
        with open(file, 'r') as f:
            data = json.load(f)
            # if the file is not from a self play episode (M-FOS randomly decides whether to play self or against opponent)
            if 'p1_self' not in data:
                continue
                
            # extract the data to our liking
            clean_data = {
                'mean_rewards_0': data['reward 0']/MFOS_NUM_STEPS,
                'mean_rewards_1': data['reward 1']/MFOS_NUM_STEPS,
                'agent_0_reward_against_ad': data['ad_reward_agent_0_against_always_defect']/MFOS_NUM_STEPS,
                'step': i,
                'walltime': data['walltime'],
                
            }
            metrics.append(clean_data)
    Run = {'metrics': metrics, 'size': grid_size, 'algo_name': 'M-FOS', 'seed': seed, 'wandb_run': 'mfos'} # I don't have the wandb run, but we put it so the code use it in the same way
    return Run
```

```python
def normalize(size, m):
    return m*(size//2)*2

def threshold(run, metric, eor_min: float, aar_min: float,  log: bool = False):
    size = run['size']
    each_other_reward = (metric['mean_rewards_0'] + metric['mean_rewards_1'])/2 
    #renormalize
    each_other_reward = normalize(size, each_other_reward)
    
    against_ad_reward = metric['agent_0_reward_against_ad'] 
    #renormalize
    against_ad_reward = normalize(size, against_ad_reward)
    
    if log:
        print(f'each_other_reward: {each_other_reward}')
        print(f'against_ad_reward: {against_ad_reward}')
        print(f'size: {size}')
    
    data = {'each_other_reward': each_other_reward, 'against_ad_reward': against_ad_reward, 'size': size}
    if each_other_reward > eor_min and against_ad_reward > aar_min:
        acc = True
    else:
        acc = False
    
    return {'acc': acc, 'data': data}
```

```python
import pandas as pd
# make a dataframe with columns size algo step walltime

def generate_pass_df(eor_min, aar_min):
    df = pd.DataFrame(columns=['size', 'algo', 'step', 'walltime', 'eor', 'aar'])
    
    for Run in Runs:
        run_name = str(Run['wandb_run'])
        if 'loqa-ipd' in run_name:
            algo = 'LOQA-IPD'
        elif 'Pola' in run_name:
            algo = 'POLA'
        elif 'mfos' in run_name:
            algo = 'M-FOS'
        else:
            raise ValueError('No name found')
        data = get_first_threshold_pass(lambda r, m, log:threshold(r, m, eor_min=eor_min, aar_min=aar_min, log=log), Run)
        df.loc[len(df)] = {'size': Run['size'], 
                        'algo': algo, 
                        'step': data['step'], 
                        'walltime': data['walltime'],
                        }
    
    return df
```
